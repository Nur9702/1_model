##Model quramiz

```python
from fastai.vision.all import *
from ipywidgets import widgets

#path
path = Path('train')
#path.ls()
#fls = get_image_files(path)
#fls
#failed = verify_images(fls)
#failed

# Datablockni yaratib olamiz
transports = DataBlock(
    blocks=(ImageBlock, CategoryBlock),
    get_items=get_image_files, # Use robust get_items function
    splitter=RandomSplitter(valid_pct=0.2, seed=42), # datani randomly bo'ladi
    get_y=parent_label, # rasmlarga o'zini nomini qo'yib beradi
    item_tfms=Resize(224)
)
#Dataloaderni yaratamiz
dls = transports.dataloaders(path)

#Datani tekshiramiz
#dls.valid.show_batch(max_n=32, nrows=4) # 32 rasmni 4 qator qilib ko'rsatib beradi

# O'qitish (train)
learn = cnn_learner(dls, resnet34, metrics = accuracy)
learn.fine_tune(4)
```

```python
interp = ClassificationInterpretation.from_learner(learn)
interp.plot_confusion_matrix()
```

```python
interp.plot_top_losses(5, nrows=1)
```

```python
upload = widgets.FileUpload()
upload
```

```python
img = PILImage.create(upload.data[-1])
pred, pred_id, probs = learn.predict(img)
print(f'Bashorat: {pred}')
print(f'Ehtimollik: {probs[pred_id]*100:.1f}')
img
```
